# Data Cleansing

```python
#Delete unnecessary data column
penduduk_df.drop(['no', 'rt', 'rw'], axis=1, inplace=True)
```

```python
#generate number for "*" character on NIK
def generate_NIK(data):
  unique = 100000
  i=0
  for row in data['nik']:
    row = re.sub("\-", "", row)
    row = re.sub("\*", "", row)
    data.loc[i, "nik"] = str(row) + str(unique)
    i+=1
    unique +=1
  
  return data

penduduk_df = generate_NIK(penduduk_df)
```
